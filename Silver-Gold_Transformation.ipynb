```
#Read the data from the silver container 
input_path = "/mnt/silver/etl/car_price/"

delta_spark_DF = spark. \
    read. \
        format('delta').load(input_path)
```

```
display(delta_spark_DF)
```

```
#Capitalize the first letter of every word in the column name(e.g,fuel_type => Fuel_Type)
column_names = delta_spark_DF.columns

for old_column_name in column_names:
    new_column_name = '_'.join(map(lambda s: s.strip().capitalize(), old_column_name.split('_')))
    delta_spark_DF = delta_spark_DF.withColumnRenamed(old_column_name, new_column_name)
    
```

```
display(delta_spark_DF)
```

```
#Writing the Transformed Data to the Gold layer, which is the final layer.

output_path = "/mnt/gold/etl/car_price"
delta_spark_DF.write.\
    format('delta'). \
        mode('overwrite').save(output_path)
```
